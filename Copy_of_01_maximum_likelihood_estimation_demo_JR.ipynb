### Likelihood Function for Poisson Distribution:

```python
import math

def poisson_likelihood(lam, data):
    """
    Calculates the likelihood of observing the given data under a Poisson distribution
    with parameter lam (rate).

    Args:
        lam (float): Rate parameter of the Poisson distribution
        data (list or np.array): Observed data points (non-negative integers)

    Returns:
        float: Likelihood value
    """
    likelihood = 1
    for x in data:
        likelihood *= (lam**x * math.exp(-lam)) / math.factorial(x)
    return likelihood
```

### Likelihood Function for Gaussian (Normal) Distribution:

```python
import math

def gaussian_likelihood(mu, sigma, data):
    """
    Calculates the likelihood of observing the given data under a normal distribution
    with parameters mu (mean) and sigma (standard deviation).

    Args:
        mu (float): Mean of the normal distribution
        sigma (float): Standard deviation of the normal distribution
        data (list or np.array): Observed data points

    Returns:
        float: Likelihood value
    """
    n = len(data)
    likelihood = 1
    for x in data:
        likelihood *= (1 / (sigma * math.sqrt(2 * math.pi))) * math.exp(-(x - mu)**2 / (2 * sigma**2))
    return likelihood
```

```python
import numpy as np

lambda_range = np.arange(1.1, 4.6, 0.1)
likelihoods = [poisson_likelihood(lam, data) for lam in lambda_range]
```

```python
def poisson_log_likelihood(lam, data):
    return sum(x * np.log(lam) - lam - math.lgamma(x+1) for x in data)

log_likelihoods = [poisson_log_likelihood(lam, data) for lam in lambda_range]
```

3. Plot the likelihood function over a range of 𝜆 values.

```python
import matplotlib.pyplot as plt

plt.plot(lambda_range, log_likelihoods)
plt.xlabel("λ (mean incidence per day)")
plt.ylabel("Log-Likelihood")
plt.title("Poisson Likelihood for Daily Cases")
plt.grid(True)
plt.show()
```

```python
max_index = np.argmax(log_likelihoods)
lambda_mle = lambda_range[max_index]
lambda_mle
```
